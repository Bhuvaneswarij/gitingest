```python

!pip install --quiet keras-tuner statsmodels shap

# ---------------------------
# 1) Imports & seeds
# ---------------------------
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import statsmodels.api as sm
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")

# Correct Keras Tuner import (after pip install)
import keras_tuner as kt

# Import SHAP (may require a compatible TF version)
import shap

# reproducibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------------------
# 2) Create a multivariate synthetic dataset
# ---------------------------
T = 500
t = np.arange(T)

trend = 0.0008 * t
seasonal1 = 0.8 * np.sin(0.02 * t)
seasonal2 = 0.4 * np.sin(0.05 * t)
exog_indicator = ((t % 50) < 5).astype(float)  # periodic short events
noise = 0.1 * np.random.randn(T)

target = seasonal1 + seasonal2 + trend + 0.8 * exog_indicator + noise

moving_avg = pd.Series(target).rolling(window=5, min_periods=1).mean().values
hour_of_day = (t % 24) / 23.0

df = pd.DataFrame({
    "target": target,
    "seasonal2": seasonal2,
    "moving_avg": moving_avg,
    "exog_indicator": exog_indicator,
    "hour_of_day": hour_of_day
})

print("Data sample:\n", df.head())

# ---------------------------
# 3) Preprocessing: scalers + sequences
# ---------------------------
SEQ_LEN = 24
FEATURES = df.columns.tolist()

feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

scaled_all = feature_scaler.fit_transform(df.values)

def create_sequences_multivariate(arr, seq_len, target_col_index=0):
    X, y = [], []
    for i in range(len(arr)-seq_len):
        X.append(arr[i:i+seq_len, :])
        y.append(arr[i+seq_len, target_col_index])
    return np.array(X), np.array(y).reshape(-1,1)

X, y = create_sequences_multivariate(scaled_all, SEQ_LEN, target_col_index=0)
split = int(len(X)*0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

target_scaler.fit(y_train)
y_train_scaled = target_scaler.transform(y_train)
y_test_scaled = target_scaler.transform(y_test)

print("Shapes:", X_train.shape, y_train.shape, X_test.shape, y_test.shape)

# ---------------------------
# 4) LSTM builder & tuner
# ---------------------------
def build_lstm_model(hp):
    units = hp.Int('units', min_value=32, max_value=128, step=32)
    dropout = hp.Float('dropout', min_value=0.1, max_value=0.5, step=0.1)
    lr = hp.Float('lr', 1e-4, 1e-2, sampling='log')

    inp = layers.Input(shape=(SEQ_LEN, X_train.shape[2]))
    x = layers.LSTM(units, return_sequences=True)(inp)
    x = layers.Dropout(dropout)(x)
    x = layers.LSTM(units)(x)
    x = layers.Dropout(dropout)(x)
    out = layers.Dense(1)(x)
    model = models.Model(inputs=inp, outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss='mse')
    return model

tuner = kt.RandomSearch(
    build_lstm_model,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=1,
    directory='lstm_tuning',
    project_name='ts_project'
)

from tensorflow.keras.callbacks import ReduceLROnPlateau
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)

print("Starting tuning (small budget)...")
tuner.search(X_train, y_train_scaled, validation_split=0.1, epochs=15, batch_size=32, callbacks=[lr_scheduler], verbose=1)
best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()

# Retrain best model to convergence (increase epochs for real experiments)
history = best_model.fit(X_train, y_train_scaled, validation_split=0.1, epochs=30, batch_size=32, callbacks=[lr_scheduler], verbose=1)

# ---------------------------
# 5) LSTM predictions + metrics
# ---------------------------
pred_scaled = best_model.predict(X_test)
pred_inv = target_scaler.inverse_transform(pred_scaled)
y_test_inv = target_scaler.inverse_transform(y_test_scaled)

mae_l = mean_absolute_error(y_test_inv, pred_inv)
rmse_l = mean_squared_error(y_test_inv, pred_inv)**0.5
r2_l = r2_score(y_test_inv, pred_inv)
print(f"LSTM MAE={mae_l:.6f}, RMSE={rmse_l:.6f}, R2={r2_l:.6f}")

# ---------------------------
# 6) Baseline: SARIMA (univariate) and ETS
# ---------------------------
# Align raw target for baselines (account for seq_len shift used in sequence creation)
full_target = df['target'].values
train_target = full_target[:split+SEQ_LEN]
test_target = full_target[split+SEQ_LEN:]

# SARIMA quick spec (example) - in practice you'd grid-search (we keep simple)
sarima_order = (2,0,2)
seasonal_order = (1,0,1,24)
sarima_model = SARIMAX(train_target, order=sarima_order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
sarima_res = sarima_model.fit(disp=False)
sarima_pred = sarima_res.get_forecast(steps=len(test_target)).predicted_mean

mae_s = mean_absolute_error(test_target, sarima_pred)
rmse_s = mean_squared_error(test_target, sarima_pred)**0.5
r2_s = r2_score(test_target, sarima_pred)
print(f"SARIMA MAE={mae_s:.6f}, RMSE={rmse_s:.6f}, R2={r2_s:.6f}")

# Exponential Smoothing (ETS)
ets_model = ExponentialSmoothing(train_target, trend='add', seasonal='add', seasonal_periods=24)
ets_res = ets_model.fit(optimized=True)
ets_pred = ets_res.forecast(steps=len(test_target))

mae_e = mean_absolute_error(test_target, ets_pred)
rmse_e = mean_squared_error(test_target, ets_pred)**0.5
r2_e = r2_score(test_target, ets_pred)
print(f"ETS MAE={mae_e:.6f}, RMSE={rmse_e:.6f}, R2={r2_e:.6f}")

# ---------------------------
# 7) MC Dropout uncertainty
# ---------------------------
def mc_dropout_predict(model, X, n_samples=40):
    preds = []
    X_tf = tf.convert_to_tensor(X, dtype=tf.float32)
    for _ in range(n_samples):
        p = model(X_tf, training=True).numpy()
        preds.append(p)
    return np.array(preds)  # shape (n_samples, batch, 1)

samples = mc_dropout_predict(best_model, X_test, n_samples=40)
mean_pred = samples.mean(axis=0)
lower = np.percentile(samples, 5, axis=0)
upper = np.percentile(samples, 95, axis=0)

mean_pred_inv = target_scaler.inverse_transform(mean_pred)
lower_inv = target_scaler.inverse_transform(lower)
upper_inv = target_scaler.inverse_transform(upper)

# ---------------------------
# 8) SHAP explainability (DeepExplainer with fallback)
# ---------------------------
# Use small background set to speed up
bg_idx = np.random.choice(X_train.shape[0], min(50, X_train.shape[0]), replace=False)
X_train_small = X_train[bg_idx]

# Try DeepExplainer; fallback to GradientExplainer, then KernelExplainer (very small sample)
shap_vals = None
try:
    explainer = shap.DeepExplainer(best_model, X_train_small)
    to_explain = X_test[:30]
    shap_values = explainer.shap_values(to_explain)
    if isinstance(shap_values, list):
        shap_vals = shap_values[0]  # (samples, seq_len, n_features)
    else:
        shap_vals = shap_values
    print("Used DeepExplainer.")
except Exception as e:
    print("DeepExplainer failed, trying GradientExplainer:", e)
    try:
        explainer = shap.GradientExplainer(best_model, X_train_small)
        to_explain = X_test[:30]
        shap_values = explainer.shap_values(to_explain)
        if isinstance(shap_values, list):
            shap_vals = shap_values[0]
        else:
            shap_vals = shap_values
        print("Used GradientExplainer.")
    except Exception as e2:
        print("GradientExplainer failed, trying KernelExplainer fallback:", e2)
        # kernel on a very small set (slow). We'll reduce nsamples to speed up.
        explainer = shap.KernelExplainer(lambda x: best_model.predict(x.reshape(-1, SEQ_LEN, X_train.shape[2])).ravel(), X_train_small.reshape(X_train_small.shape[0], -1))
        to_explain_flat = X_test[:10].reshape(10, -1)
        shap_values_flat = explainer.shap_values(to_explain_flat, nsamples=200)
        # reshape back: (samples, seq_len, n_features)
        shap_vals = np.array(shap_values_flat).reshape(to_explain_flat.shape[0], SEQ_LEN, X_train.shape[2])
        print("Used KernelExplainer fallback (small sample).")

if shap_vals is not None:
    print("SHAP values shape:", shap_vals.shape)
    # Per-timestep importance (sum over features)
    per_timestep_importance = np.sum(np.abs(shap_vals), axis=2)  # (samples, seq_len)
    avg_time_importance = np.mean(per_timestep_importance, axis=0)  # (seq_len,)
    # Plot average importance with most recent lag on right
    plt.figure(figsize=(9,3))
    plt.plot(np.arange(-SEQ_LEN,0), avg_time_importance[::-1])
    plt.title("Average per-timestep absolute SHAP importance (most recent on right)")
    plt.xlabel("Lag (steps back)")
    plt.ylabel("Mean |SHAP|")
    plt.show()

    # Per-feature importance
    per_feature_importance = np.sum(np.abs(shap_vals), axis=1)  # (samples, n_features)
    avg_feat_imp = np.mean(per_feature_importance, axis=0)
    print("Per-feature mean |SHAP|:")
    for name, val in zip(FEATURES, avg_feat_imp):
        print(f"  {name}: {val:.6f}")
else:
    print("SHAP values not computed.")

# ---------------------------
# 9) Save results and write final report files
# ---------------------------
results_table = pd.DataFrame({
    "Model": ["LSTM", "SARIMA", "ETS"],
    "MAE": [mae_l, mae_s, mae_e],
    "RMSE": [rmse_l, rmse_s, rmse_e],
    "R2": [r2_l, r2_s, r2_e]
})
print("\nComparison table:\n", results_table)
results_table.to_csv("model_comparison.csv", index=False)

df_res = pd.DataFrame({
    "Actual": test_target,
    "LSTM_Mean": mean_pred_inv.flatten(),
    "LSTM_Lower": lower_inv.flatten(),
    "LSTM_Upper": upper_inv.flatten(),
    "SARIMA": sarima_pred,
    "ETS": ets_pred
})
df_res.to_csv("forecast_results_with_baselines.csv", index=False)
print("Saved forecast_results_with_baselines.csv")

# Write final report (README.md + FINAL_REPORT.txt)
final_report_md = f"""
# LSTM Time Series Forecasting Project — FINAL REPORT

## Executive summary
This project trains a tuned LSTM on a multivariate synthetic dataset, compares it against SARIMA and ETS baselines, provides per-timestep explainability via SHAP, and quantifies uncertainty using MC Dropout.

## Data
- 500 time steps, features: {FEATURES}
- Train/test split: 80/20
- Sequence length: {SEQ_LEN}

## Models & Metrics
- LSTM: MAE={mae_l:.6f}, RMSE={rmse_l:.6f}, R2={r2_l:.6f}
- SARIMA: MAE={mae_s:.6f}, RMSE={rmse_s:.6f}, R2={r2_s:.6f}
- ETS: MAE={mae_e:.6f}, RMSE={rmse_e:.6f}, R2={r2_e:.6f}

## Explainability
- SHAP used (DeepExplainer preferred; fallback to Gradient/Kernal if needed)
- Outputs per-timestep and per-feature attributions.

## Uncertainty
- MC Dropout (n=40) used to estimate predictive 5-95% CI for LSTM.

## Files saved
- model_comparison.csv
- forecast_results_with_baselines.csv
- README.md
"""

with open("README.md", "w") as f:
    f.write(final_report_md)

with open("FINAL_REPORT.txt", "w") as f:
    f.write(final_report_md)

print("Saved README.md and FINAL_REPORT.txt")

# ---------------------------
# 10) Quick plots for the user
# ---------------------------
plt.figure(figsize=(12,4))
plt.plot(test_target, label='Actual')
plt.plot(mean_pred_inv.flatten(), label='LSTM Mean')
plt.plot(sarima_pred, label='SARIMA')
plt.plot(ets_pred, label='ETS')
plt.fill_between(np.arange(len(mean_pred_inv)), lower_inv.flatten(), upper_inv.flatten(), alpha=0.25, label='LSTM 5-95% CI')
plt.legend()
plt.title("Actual vs LSTM vs SARIMA vs ETS")
plt.show()

# Training history plot
plt.figure(figsize=(8,3))
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title("Training History")
plt.legend()
plt.show()

print("\\nAll done — files saved in working directory.")
```
